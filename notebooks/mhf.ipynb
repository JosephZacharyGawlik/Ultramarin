# DeepLOB "Super Model" (DeepLOB Encoder + Seq2Seq Decoder)

This notebook implements the architecture inspired by the MHF paper.

**Architecture:**
1.  **Encoder:** DeepLOB (Convolutional Blocks + Inception Module + LSTM) to extract rich spatial features from the LOB.
2.  **Decoder:** Seq2Seq LSTM with Additive Attention to generate the 60-second midprice path step-by-step.


```python
import torch
from torch import nn
import numpy as np
import pandas as pd
from pathlib import Path
from torch.utils.data import Dataset, DataLoader
import random

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device:", device)

# Fix randomness for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
```

```python
# Paths and volume_to_fill
root = Path("/workspace/practical") if Path("/workspace/practical").exists() else Path.cwd()
import sys
if str(root / "src") not in sys.path:
    sys.path.append(str(root / "src"))

symbol_dir = root / "BTCUSDT"
X_path = symbol_dir / "X_train.parquet"
Y_path = symbol_dir / "y_train.parquet"
vol_path = symbol_dir / "vol_to_fill.txt"

volume_to_fill = None
if vol_path.exists():
    import re
    with open(vol_path) as f:
        m = re.search(r"([\d.]+)", f.read())
    if m:
        volume_to_fill = float(m.group(1))
print("volume_to_fill:", volume_to_fill)

FEATURE_COLS = []
for i in range(1, 6):
    FEATURE_COLS.extend([f"ask_price_{i}", f"ask_vol_{i}", f"bid_price_{i}", f"bid_vol_{i}"])
# Total 20 features: AP1, AV1, BP1, BV1, ...

target_window = 60
input_window = 5*60  # 300 seconds input
```

## 1. The Encoder: DeepLOB (Heavy)
This is the standard DeepLOB CNN-Inception encoder.

```python
class DeepLOBEncoder(nn.Module):
    def __init__(self, in_ch: int = 1):
        super().__init__()
        # Input: [B, 1, T, 20]
        # 1. Convolutional Blocks (Temporal Stride = 1 to preserve seq len)
        # Block 1: Feature Extraction
        # kernel=(1,2), stride=(1,2) -> Groups (Price, Vol) and reduces features 20 -> 10
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_ch, 32, kernel_size=(1,2), stride=(1,2)), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
        )
        # Block 2: 
        # kernel=(1,2), stride=(1,2) -> Reduces features 10 -> 5 (roughly one feature per level)
        self.conv2 = nn.Sequential(
            nn.Conv2d(32,32,kernel_size=(1,2), stride=(1,2)), nn.Tanh(), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.Tanh(), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.Tanh(), nn.BatchNorm2d(32),
        )
        # Block 3:
        # kernel=(1,5) -> Collapses remaining 5 feature dims to 1 (conceptually)
        # Output here will be [B, 32, T, 1]
        self.conv3 = nn.Sequential(
            nn.Conv2d(32,32,kernel_size=(1,5)), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
            nn.Conv2d(32,32,kernel_size=(4,1), padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(32),
        )
        
        # Inception Module (Preserves Time T, applied on the feature map)
        # Input to Inception: [B, 32, T, 1] after conv3
        self.inp1 = nn.Sequential(
            nn.Conv2d(32,64,kernel_size=(1,1),padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(64),
            nn.Conv2d(64,64,kernel_size=(3,1),padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(64),
        )
        self.inp2 = nn.Sequential(
            nn.Conv2d(32,64,kernel_size=(1,1),padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(64),
            nn.Conv2d(64,64,kernel_size=(5,1),padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(64),
        )
        self.inp3 = nn.Sequential(
            nn.MaxPool2d((3,1),stride=(1,1),padding=(1,0)),
            nn.Conv2d(32,64,kernel_size=(1,1),padding='same'), nn.LeakyReLU(0.01), nn.BatchNorm2d(64),
        )
        
    def forward(self, x):
        # x: [B, T, F] -> [B, 1, T, 20]
        # We treat LOB as an image with 1 channel and 20 features height.
        x = x.unsqueeze(1)
        
        x = self.conv1(x) 
        x = self.conv2(x) 
        x = self.conv3(x) 
        
        # Inception
        x_inp = torch.cat([self.inp1(x), self.inp2(x), self.inp3(x)], dim=1) # [B, 192, T, 1]
        
        # Reshape to [B, T, Features] for LSTM
        # Permute: [B, 192, T, 1] -> [B, T, 192, 1] -> [B, T, 192]
        x_out = x_inp.permute(0, 2, 1, 3).squeeze(-1)
        return x_out
```

## 2. The Decoder: Seq2Seq Attention
This replaces the simple Bi-LSTM head with a full autoregressive decoder.

```python
class AdditiveAttention(nn.Module):
    def __init__(self, enc_dim: int, dec_dim: int, attn_dim: int = 64):
        super().__init__()
        self.W_e = nn.Linear(enc_dim, attn_dim, bias=False)
        self.W_d = nn.Linear(dec_dim, attn_dim, bias=False)
        self.v = nn.Linear(attn_dim, 1, bias=False)

    def forward(self, enc_outputs: torch.Tensor, dec_state: torch.Tensor) -> torch.Tensor:
        # enc_outputs: [B, T, E], dec_state: [B, D]
        e = self.W_e(enc_outputs) + self.W_d(dec_state).unsqueeze(1)  # [B, T, A]
        scores = self.v(torch.tanh(e)).squeeze(-1)  # [B, T]
        weights = torch.softmax(scores, dim=1)
        context = torch.sum(weights.unsqueeze(-1) * enc_outputs, dim=1)  # [B, E]
        return context

class SuperModel(nn.Module):
    def __init__(self, hidden: int = 128, horizon: int = 60, dropout: float=0.1):
        super().__init__()
        # 1. Encoder Part
        self.spatial = DeepLOBEncoder(in_ch=1) # The Heavy CNN [B, T, 192]
        self.encoder = nn.LSTM(
            input_size=192, # DeepLOB output dim
            hidden_size=hidden,
            num_layers=1,
            batch_first=True,
            bidirectional=True
        )
        
        # 2. Decoder Part
        self.attn = AdditiveAttention(enc_dim=2 * hidden, dec_dim=hidden)
        self.decoder = nn.LSTM(
            input_size=1 + 2 * hidden,  # previous y + context
            hidden_size=hidden,
            num_layers=1,
            batch_first=True,
            dropout=0.0,
        )
        self.out = nn.Linear(hidden, 1)
        self.init_h = nn.Linear(2 * hidden, hidden)
        self.init_c = nn.Linear(2 * hidden, hidden)
        self.horizon = horizon

    def forward(self, x: torch.Tensor, y_teacher: torch.Tensor = None) -> torch.Tensor:
        # x: [B, T, F]
        # y_teacher: [B, 60]
        
        # Encode
        h_spatial = self.spatial(x)  # [B, T, 192]
        enc_out, _ = self.encoder(h_spatial)  # [B, T, 2H]

        # Init decoder state from encoder mean
        enc_mean = enc_out.mean(dim=1)
        dec_h0 = torch.tanh(self.init_h(enc_mean))
        dec_c0 = torch.tanh(self.init_c(enc_mean))
        dec_h = dec_h0.unsqueeze(0)  # [1, B, H]
        dec_c = dec_c0.unsqueeze(0)

        outputs = []
        # Seed: use last encoder input midprice proxy (col 0 normalized)
        prev_y = x[:, -1, 0].unsqueeze(-1) # [B, 1]

        for t in range(self.horizon):
            context = self.attn(enc_out, dec_h[-1])  # [B, 2H]
            dec_in = torch.cat([prev_y, context], dim=1).unsqueeze(1)  # [B, 1, 1+2H]
            dec_out, (dec_h, dec_c) = self.decoder(dec_in, (dec_h, dec_c))
            y_hat = self.out(dec_out).squeeze(1)  # [B, 1]
            outputs.append(y_hat.squeeze(-1))
            if self.training and y_teacher is not None:
                # Teacher forcing
                prev_y = y_teacher[:, t].unsqueeze(-1)
            else:
                # Autoregressive
                prev_y = y_hat.detach()

        return torch.stack(outputs, dim=1)  # [B, 60]

def forecast_loss(pred, target, smooth_lambda=0.02):
    mse = (pred - target) ** 2
    loss = mse.mean()
    if smooth_lambda > 0:
        smooth = (pred[:, 1:] - pred[:, :-1]) ** 2
        loss = loss + smooth_lambda * smooth.mean()
    return loss
```

## 3. Dataset & Training Loop

```python
# Normalized dataset
class LastMinuteDatasetNorm(Dataset):
    def __init__(self, x_df: pd.DataFrame, y_df: pd.DataFrame, feat_means, feat_stds, mid_mean, mid_std):
        self.samples = []
        for hid, xh in x_df.groupby("anonymized_id"):
            yh = y_df[y_df["anonymized_id"] == hid]
            if len(yh) != target_window:
                continue
            xh = xh.sort_values("time_in_hour").tail(input_window)
            x_arr = xh[FEATURE_COLS].astype(np.float32).to_numpy()
            if x_arr.shape[0] < input_window:
                pad = np.zeros((input_window - x_arr.shape[0], x_arr.shape[1]), dtype=np.float32)
                x_arr = np.vstack([pad, x_arr])
            mid = ((yh["ask_price_1"] + yh["bid_price_1"]) / 2.0).astype(np.float32).to_numpy()
            if mid.shape[0] != target_window:
                continue
            x_arr = (x_arr - feat_means) / feat_stds
            mid_norm = (mid - mid_mean) / mid_std
            if not np.isfinite(x_arr).all() or not np.isfinite(mid_norm).all():
                continue
            self.samples.append((x_arr, mid_norm))
    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        x, y = self.samples[idx]
        return torch.from_numpy(x), torch.from_numpy(y)

from dataclasses import dataclass

def chrono_split(x_df, y_df, val_ratio=0.2):
    ids = np.sort(x_df["anonymized_id"].unique())
    split = int(len(ids)*(1-val_ratio))
    tr_ids, va_ids = ids[:split], ids[split:]
    x_tr = x_df[x_df["anonymized_id"].isin(tr_ids)].reset_index(drop=True)
    x_va = x_df[x_df["anonymized_id"].isin(va_ids)].reset_index(drop=True)
    y_tr = y_df[y_df["anonymized_id"].isin(tr_ids)].reset_index(drop=True)
    y_va = y_df[y_df["anonymized_id"].isin(va_ids)].reset_index(drop=True)
    return x_tr, x_va, y_tr, y_va

@dataclass
class TrainCfg:
    epochs: int = 15
    batch_size: int = 16
    lr: float = 1e-3
    weight_decay: float = 1e-5
    smooth_lambda: float = 0.02
    val_ratio: float = 0.2

def train_val():
    X = pd.read_parquet(X_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    Y = pd.read_parquet(Y_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    x_tr, x_va, y_tr, y_va = chrono_split(X, Y, val_ratio=TrainCfg.val_ratio)

    feat_means = x_tr[FEATURE_COLS].mean().to_numpy().astype(np.float32)
    feat_stds  = x_tr[FEATURE_COLS].std().replace(0,1e-6).to_numpy().astype(np.float32)
    mid_tr = (y_tr["ask_price_1"] + y_tr["bid_price_1"]) / 2.0
    mid_mean = float(mid_tr.mean()); mid_std = float(mid_tr.std() if mid_tr.std()!=0 else 1e-6)

    train_ds = LastMinuteDatasetNorm(x_tr, y_tr, feat_means, feat_stds, mid_mean, mid_std)
    val_ds   = LastMinuteDatasetNorm(x_va, y_va, feat_means, feat_stds, mid_mean, mid_std)
    train_loader = DataLoader(train_ds, batch_size=TrainCfg.batch_size, shuffle=True)
    val_loader   = DataLoader(val_ds, batch_size=TrainCfg.batch_size, shuffle=False)

    model = SuperModel().to(device)
    opt = torch.optim.Adam(model.parameters(), lr=TrainCfg.lr, weight_decay=TrainCfg.weight_decay)

    for epoch in range(TrainCfg.epochs):
        tr = train_epoch(model, train_loader, opt, TrainCfg)
        va = eval_epoch(model, val_loader, TrainCfg)
        print(f"epoch {epoch+1}: train={tr:.4f} val={va:.4f}")

    scalers = {"feat_means": feat_means, "feat_stds": feat_stds, "mid_mean": mid_mean, "mid_std": mid_std}
    return model, scalers, val_loader

def train_epoch(model, loader, opt, cfg: TrainCfg):
    model.train(); total=0; n=0
    for xb, yb in loader:
        xb = xb.to(device); yb = yb.to(device)
        opt.zero_grad()
        # Pass yb as teacher
        pred = model(xb, y_teacher=yb)
        loss = forecast_loss(pred, yb, cfg.smooth_lambda)
        loss.backward(); opt.step()
        total += float(loss.item())*xb.size(0); n += xb.size(0)
    return total/max(n,1)

def eval_epoch(model, loader, cfg: TrainCfg):
    model.eval(); total=0; n=0
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device); yb = yb.to(device)
            # No teacher forcing during eval
            pred = model(xb, y_teacher=None)
            loss = forecast_loss(pred, yb, cfg.smooth_lambda)
            total += float(loss.item())*xb.size(0); n += xb.size(0)
    return total/max(n,1)
```

```python
# Train
run_train = True
if run_train and X_path.exists() and Y_path.exists():
    model, scalers, val_loader = train_val()
else:
    print("Set run_train=True and ensure X_path/Y_path exist.")
```

## 4. Evaluation
Calculate implementation error (bps) usage `alpha=1.0` (pure model).

```python
from simulate_walk_the_book import simulate_walk_the_book
try:
    from predictive_scheduler import schedule_from_forecasts
except ImportError:
    from predictive_scheduler import build_schedule_from_forecasts as schedule_from_forecasts

run_eval = True
if run_eval and X_path.exists() and Y_path.exists() and 'model' in globals():
    X = pd.read_parquet(X_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    Y = pd.read_parquet(Y_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    ids = np.sort(X["anonymized_id"].unique()); split = int(len(ids)*0.8); va_ids = ids[split:]
    x_va_raw = X[X["anonymized_id"].isin(va_ids)].reset_index(drop=True)
    y_va_raw = Y[Y["anonymized_id"].isin(va_ids)].reset_index(drop=True)

    errs = []
    model.eval()
    for hid in va_ids:
        xh_raw = x_va_raw[x_va_raw["anonymized_id"]==hid].sort_values("time_in_hour")
        yh_raw = y_va_raw[y_va_raw["anonymized_id"]==hid].sort_values("time_in_hour")
        if len(yh_raw)!=target_window:
            continue
        # normalize input
        feat_means = scalers['feat_means']; feat_stds = scalers['feat_stds']; mid_mean = scalers['mid_mean']; mid_std = scalers['mid_std']
        xh = xh_raw.tail(input_window)
        # Pad if needed
        x_arr = xh[FEATURE_COLS].astype(np.float32).to_numpy()
        if x_arr.shape[0] < input_window:
            pad = np.zeros((input_window - x_arr.shape[0], x_arr.shape[1]), dtype=np.float32)
            x_arr = np.vstack([pad, x_arr])
        x_arr = (x_arr - feat_means) / feat_stds
        xb = torch.from_numpy(x_arr).unsqueeze(0).to(device)
        with torch.no_grad():
            # No teacher forcing during eval
            pred = model(xb, y_teacher=None)
        mid_pred = (pred.cpu().numpy()[0] * mid_std) + mid_mean

        # Build positions with alpha=1.0
        def end_window_twap(vol, window=14):
            w = np.zeros(60, dtype=np.float32); w[-window:] = vol / window; return w
        sched_forecast = schedule_from_forecasts(mid_pred, volume_to_fill)
        sched_twap = end_window_twap(volume_to_fill, window=14)
        alpha = 1.0  # FULL MODEL CONFIDENCE
        positions = alpha * sched_forecast + (1-alpha) * sched_twap
        positions = positions * (volume_to_fill / positions.sum()) if positions.sum()!=0 else sched_twap

        ask_prices = np.column_stack([yh_raw[f"ask_price_{lvl}"].to_numpy(dtype=float) for lvl in range(1,6)])
        ask_vols   = np.column_stack([yh_raw[f"ask_vol_{lvl}"].to_numpy(dtype=float)   for lvl in range(1,6)])
        bid_prices = np.column_stack([yh_raw[f"bid_price_{lvl}"].to_numpy(dtype=float) for lvl in range(1,6)])
        bid_vols   = np.column_stack([yh_raw[f"bid_vol_{lvl}"].to_numpy(dtype=float)   for lvl in range(1,6)])
        total_vol, avg_price = simulate_walk_the_book(positions, ask_prices, ask_vols, bid_prices, bid_vols)
        close_price = yh_raw["close"].dropna().iloc[-1] if len(yh_raw["close"].dropna())>0 else np.nan
        if total_vol>0 and not np.isnan(avg_price) and not np.isnan(close_price):
            rel_err = abs(avg_price - close_price) / close_price
            fill_penalty = min(100.0, volume_to_fill / total_vol)
            err_bps = rel_err * fill_penalty * 10000.0
            errs.append(err_bps)
    if errs:
        print(f"Validation implementation error (bps) over {len(errs)} hours: mean={np.mean(errs):.4f}, median={np.median(errs):.4f}")
    else:
        print("No valid errors computed.")
else:
    print("Set run_eval=True after training to evaluate.")
```

```python
# === CLEANED UP EVALUATION LOOP ===
# This cell performs the exact same logic as above but with clearer steps and variable names.

if run_eval and X_path.exists() and Y_path.exists() and 'model' in globals():
    print("Starting Clean Evaluation...")
    
    # 1. Load Validation Data
    # We reload and split just to be safe and explicit
    X_full = pd.read_parquet(X_path).sort_values(["anonymized_id", "time_in_hour"]).reset_index(drop=True)
    Y_full = pd.read_parquet(Y_path).sort_values(["anonymized_id", "time_in_hour"]).reset_index(drop=True)
    
    # Split by ID (Chronological Split)
    unique_ids = np.sort(X_full["anonymized_id"].unique())
    split_idx = int(len(unique_ids) * 0.8)
    validation_ids = unique_ids[split_idx:]
    
    # Filter dataframes for validation set only
    X_val = X_full[X_full["anonymized_id"].isin(validation_ids)].reset_index(drop=True)
    Y_val = Y_full[Y_full["anonymized_id"].isin(validation_ids)].reset_index(drop=True)
    
    # Retrieve Training Scalers (Must use same stats as training!)
    feature_means = scalers['feat_means']
    feature_stds  = scalers['feat_stds']
    midprice_mean = scalers['mid_mean']
    midprice_std  = scalers['mid_std']

    implementation_shortfalls = []
    model.eval()
    
    # 2. Iterate Over Each Hour in Validation Set
    for hour_id in validation_ids:
        # Extract data for this specific hour
        X_hour = X_val[X_val["anonymized_id"] == hour_id].sort_values("time_in_hour")
        Y_hour = Y_val[Y_val["anonymized_id"] == hour_id].sort_values("time_in_hour")
        
        # Skip incomplete hours (must have full target window)
        if len(Y_hour) != target_window:
            continue
            
        # 3. Prepare Input for Model
        # Take the last 'input_window' seconds (e.g. 100s or 300s)
        X_input_raw = X_hour.tail(input_window)[FEATURE_COLS].astype(np.float32).to_numpy()
        
        # Handle Padding if input is too short (rare)
        current_length = X_input_raw.shape[0]
        if current_length < input_window:
            padding = np.zeros((input_window - current_length, X_input_raw.shape[1]), dtype=np.float32)
            X_input_raw = np.vstack([padding, X_input_raw])
            
        # Normalize Input (Z-Score)
        X_normalized = (X_input_raw - feature_means) / feature_stds
        
        # Convert to Tensor [1, T, F]
        X_tensor = torch.from_numpy(X_normalized).unsqueeze(0).to(device)
        
        # 4. Run Model Prediction
        with torch.no_grad():
            # y_teacher=None forces autoregressive mode (predicting step-by-step)
            prediction_normalized = model(X_tensor, y_teacher=None)
            
        # De-normalize Prediction to get Real Prices
        midprice_forecast = (prediction_normalized.cpu().numpy()[0] * midprice_std) + midprice_mean
        
        # 5. Generate Execution Schedule
        # "Smart" Schedule based on forecast
        smart_schedule = schedule_from_forecasts(midprice_forecast, volume_to_fill)
        
        # "Dumb" Schedule (TWAP) as fallback/baseline component
        # Spreads volume evenly over the last 14 seconds
        twap_schedule = np.zeros(60, dtype=np.float32)
        twap_schedule[-14:] = volume_to_fill / 14
        
        # Combine Strategies (Alpha=1.0 means Pure Smart Strategy)
        alpha = 1.0
        final_positions = alpha * smart_schedule + (1 - alpha) * twap_schedule
        
        # Ensure total volume sums strictly to volume_to_fill
        total_scheduled = final_positions.sum()
        if total_scheduled > 0:
            final_positions = final_positions * (volume_to_fill / total_scheduled)
        else:
            final_positions = twap_schedule # Fallback if model outputs zero volume

        # 6. Simulate Execution in Real Market
        # Extract arrays for simulation
        ask_prices = np.column_stack([Y_hour[f"ask_price_{i}"].to_numpy(dtype=float) for i in range(1, 6)])
        ask_volumes = np.column_stack([Y_hour[f"ask_vol_{i}"].to_numpy(dtype=float)   for i in range(1, 6)])
        bid_prices = np.column_stack([Y_hour[f"bid_price_{i}"].to_numpy(dtype=float) for i in range(1, 6)])
        bid_volumes = np.column_stack([Y_hour[f"bid_vol_{i}"].to_numpy(dtype=float)   for i in range(1, 6)])
        
        executed_vol, avg_execution_price = simulate_walk_the_book(
            final_positions, 
            ask_prices, 
            ask_volumes, 
            bid_prices, 
            bid_volumes
        )
        
        # 7. Calculate Error (Basis Points)
        # Bechmark: Price at the very end of the minute
        benchmark_price = Y_hour["close"].dropna().iloc[-1] if len(Y_hour["close"].dropna()) > 0 else np.nan
        
        if executed_vol > 0 and not np.isnan(avg_execution_price) and not np.isnan(benchmark_price):
            # Relative Error
            relative_error = abs(avg_execution_price - benchmark_price) / benchmark_price
            
            # Penalty factor if we failed to fill 100% of volume
            fill_penalty = min(100.0, volume_to_fill / executed_vol)
            
            # Final Error in Basis Points (bps)
            error_bps = relative_error * fill_penalty * 10000.0
            implementation_shortfalls.append(error_bps)
            
    # 8. Report Results
    if implementation_shortfalls:
        mean_bps = np.mean(implementation_shortfalls)
        median_bps = np.median(implementation_shortfalls)
        print(f"\n=== Validation Results ===")
        print(f"Number of hours evaluated: {len(implementation_shortfalls)}")
        print(f"Mean Implementation Shortfall:   {mean_bps:.4f} bps")
        print(f"Median Implementation Shortfall: {median_bps:.4f} bps")
    else:
        print("No valid errors computed (check data or paths).")
```

```python
# R^2 Score Calculation
run_r2 = True
if run_r2 and X_path.exists() and Y_path.exists() and 'model' in globals():
    from sklearn.metrics import r2_score
    
    X = pd.read_parquet(X_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    Y = pd.read_parquet(Y_path).sort_values(["anonymized_id","time_in_hour"]).reset_index(drop=True)
    ids = np.sort(X["anonymized_id"].unique()); split = int(len(ids)*0.8); va_ids = ids[split:]
    x_va_raw = X[X["anonymized_id"].isin(va_ids)].reset_index(drop=True)
    y_va_raw = Y[Y["anonymized_id"].isin(va_ids)].reset_index(drop=True)

    all_preds = []
    all_targets = []

    model.eval()
    with torch.no_grad():
        for hid in va_ids:
            xh_raw = x_va_raw[x_va_raw["anonymized_id"]==hid].sort_values("time_in_hour")
            yh_raw = y_va_raw[y_va_raw["anonymized_id"]==hid].sort_values("time_in_hour")
            if len(yh_raw)!=target_window:
                continue
            
            # Get Ground Truth Midprice
            mid_true = ((yh_raw["ask_price_1"] + yh_raw["bid_price_1"]) / 2.0).to_numpy()

            # Prepare Input
            feat_means = scalers['feat_means']; feat_stds = scalers['feat_stds']; mid_mean = scalers['mid_mean']; mid_std = scalers['mid_std']
            xh = xh_raw.tail(input_window)
            x_arr = xh[FEATURE_COLS].astype(np.float32).to_numpy()
            if x_arr.shape[0] < input_window:
                pad = np.zeros((input_window - x_arr.shape[0], x_arr.shape[1]), dtype=np.float32)
                x_arr = np.vstack([pad, x_arr])
            x_arr = (x_arr - feat_means) / feat_stds
            xb = torch.from_numpy(x_arr).unsqueeze(0).to(device)
            
            # Predict
            # No teacher forcing during eval
            pred_norm = model(xb, y_teacher=None)
            mid_pred = (pred_norm.cpu().numpy()[0] * mid_std) + mid_mean
            
            all_preds.append(mid_pred)
            all_targets.append(mid_true)
    
    if all_preds:
        # Flatten arrays: we care about R2 over all predicted seconds
        flat_preds = np.concatenate(all_preds)
        flat_targets = np.concatenate(all_targets)
        
        # MASK NANS to avoid ValueError
        mask = np.isfinite(flat_preds) & np.isfinite(flat_targets)
        valid_preds = flat_preds[mask]
        valid_targets = flat_targets[mask]
        
        if len(valid_preds) > 0:
            r2 = r2_score(valid_targets, valid_preds)
            print(f"Validation R^2 Score (all 60s steps): {r2:.4f}")
            print(f"(Evaluated on {len(valid_preds)} valid points, {len(flat_preds)-len(valid_preds)} NaNs removed)")
        else:
            print("All predictions/targets were NaN or missing.")

    else:
        print("No predictions made.")
else:
    print("Set run_r2=True to evaluate.")
```

```python
# === MSE & Loss Evaluation ===
print("Calculating Validation Loss & MSE...")

# Ensure val_loader is ready
if 'val_loader' not in globals() or 'model' not in globals():
    print("Warning: val_loader or model not found in globals. Make sure you ran the Training cells first.")
else:
    model.eval()
    total_loss = 0
    total_mse = 0
    total_samples = 0
    
    # Use the same config as training for consistency
    curr_smooth_lambda = 0.02 # Default from TrainCfg
    if 'TrainCfg' in globals():
        curr_smooth_lambda = TrainCfg.smooth_lambda
    
    with torch.no_grad():
        for xb, yb in val_loader:
            xb = xb.to(device)
            yb = yb.to(device)
            
            # Predict (Autoregressive)
            pred = model(xb, y_teacher=None)
            
            # 1. Loss (MSE + Smoothness as defined in forecast_loss)
            loss = forecast_loss(pred, yb, smooth_lambda=curr_smooth_lambda)
            
            # 2. Pure MSE (Normalized scale)
            # pred and yb are both normalized Z-scores
            mse = ((pred - yb) ** 2).mean()
            
            batch_size = xb.size(0)
            total_loss += float(loss.item()) * batch_size
            total_mse += float(mse.item()) * batch_size
            total_samples += batch_size
    
    avg_loss = total_loss / max(total_samples, 1)
    avg_mse = total_mse / max(total_samples, 1)
    
    print(f"Validation Loss (MSE + Smooth={curr_smooth_lambda}): {avg_loss:.6f}")
    print(f"Validation MSE (Normalized):           {avg_mse:.6f}")
    print(f"Validation RMSE (Normalized):          {np.sqrt(avg_mse):.6f}")
```

```python
# === TEST SET INFERENCE ===
# Generate positions for X_test to be evaluated externally.

X_test_path = symbol_dir / "X_test.parquet"
output_path = symbol_dir / "test_positions.parquet"

if X_test_path.exists() and 'model' in globals():
    print(f"Found X_test at {X_test_path}. Generating positions...")
    
    # 1. Load Test Data
    X_test = pd.read_parquet(X_test_path).sort_values(["anonymized_id", "time_in_hour"]).reset_index(drop=True)
    unique_ids = np.unique(X_test["anonymized_id"])
    
    # Scalers
    feature_means = scalers['feat_means']
    feature_stds  = scalers['feat_stds']
    midprice_mean = scalers['mid_mean']
    midprice_std  = scalers['mid_std']
    
    results = []
    model.eval()
    
    for hour_id in unique_ids:
        # Extract hour
        X_hour = X_test[X_test["anonymized_id"] == hour_id].sort_values("time_in_hour")
        
        # 2. Prepare Input
        # Take last 'input_window' seconds
        X_input_raw = X_hour.tail(input_window)[FEATURE_COLS].astype(np.float32).to_numpy()
        
        # Pad if needed
        current_length = X_input_raw.shape[0]
        if current_length < input_window:
            padding = np.zeros((input_window - current_length, X_input_raw.shape[1]), dtype=np.float32)
            X_input_raw = np.vstack([padding, X_input_raw])
            
        # Normalize
        X_normalized = (X_input_raw - feature_means) / feature_stds
        X_tensor = torch.from_numpy(X_normalized).unsqueeze(0).to(device)
        
        # 3. Predict Midprice Path
        with torch.no_grad():
            pred_norm = model(X_tensor, y_teacher=None)
        midprice_forecast = (pred_norm.cpu().numpy()[0] * midprice_std) + midprice_mean
        
        # 4. Generate Positions (Alpha=1.0)
        smart_schedule = schedule_from_forecasts(midprice_forecast, volume_to_fill)
        
        # TWAP Fallback
        twap_schedule = np.zeros(60, dtype=np.float32)
        twap_schedule[-14:] = volume_to_fill / 14
        
        # Final Schedule
        alpha = 1.0
        final_positions = alpha * smart_schedule + (1 - alpha) * twap_schedule
        
        # Normalize to volume_to_fill
        if final_positions.sum() > 0:
            final_positions = final_positions * (volume_to_fill / final_positions.sum())
        else:
            final_positions = twap_schedule
            
        # Store Result
        # We store the ID and the array of 60 positions
        results.append({
            "anonymized_id": hour_id,
            "positions": final_positions.tolist(),
            "midprice_forecast": midprice_forecast.tolist()
        })
        
    # 5. Save to Parquet
    if results:
        submission_df = pd.DataFrame(results)
        submission_df.to_parquet(output_path)
        print(f"Successfully generated positions for {len(results)} hours.")
        print(f"Saved to: {output_path}")
        print("Columns: anonymized_id, positions (list), midprice_forecast (list)")
    else:
        print("No results generated.")
        
else:
    print(f"Skipping Test Inference: X_test not found at {X_test_path} or model not trained.")
```
